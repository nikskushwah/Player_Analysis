```python
def Appear_ratio():
 Appear = sanchez8["Team"].count()
 Goals = sanchez8["Goals"].sum()
 ratio = Goals/Appear
 print("Total Appearances: " +str(Appear))
 print("Total Goals: " +str(Goals))
 print("Goals to Appearance ratio: " +str(ratio))
 
 return
```

```python
def ShotConversion():
    Goals = sanchez8['Goals'].sum()
    Shots = sanchez8['Shots'].sum()
    Convert = Goals/Shots
    print('Total goals: ' +str(Goals))
    print('Total shots: ' +str(Shots))
    print('Shot conversion rate: ' +str(Convert))
    
    return
```

```python
Ground = []
for i in sanchez8["Opponent"]:
 Ground.append(i[0])
sanchez8["Ground"] = Ground
Month = []
for i in sanchez8["Date"]:
 Month.append(i[5:7])
sanchez8["Month"] = Month
```

```python
sanchez8
```

```python
def plotGoal_Month():
    Aug = sanchez8.loc[sanchez8["Month"] == "08", "Goals"].sum()
    Sep = sanchez8.loc[sanchez8["Month"] == "09", "Goals"].sum()
    Oct = sanchez8.loc[sanchez8["Month"] == "10", "Goals"].sum()
    Nov = sanchez8.loc[sanchez8["Month"] == "11", "Goals"].sum()
    Dec = sanchez8.loc[sanchez8["Month"] == "12", "Goals"].sum()
    Jan = sanchez8.loc[sanchez8["Month"] == "01", "Goals"].sum()
    Feb = sanchez8.loc[sanchez8["Month"] == "02", "Goals"].sum()
    Mar = sanchez8.loc[sanchez8["Month"] == "03", "Goals"].sum()
    Apr = sanchez8.loc[sanchez8["Month"] == "04", "Goals"].sum()
    May = sanchez8.loc[sanchez8["Month"] == "05", "Goals"].sum()
    Jun = sanchez8.loc[sanchez8["Month"] == "06", "Goals"].sum()
    Jul = sanchez8.loc[sanchez8["Month"] == "07", "Goals"].sum()
    
    GoalM = [Aug,Sep,Oct,Nov,Dec,Jan,Feb,Mar,Apr,May,Jun,Jul]
    x = ["Aug","Sep","Oct","Nov","Dec","Jan","Feb","Mar","Apr","May","Jun","Jul"]
    sb.set()
    dims = (11.7,8.27) #A4 dimensions
    fig, ax = plt.subplots(figsize=dims)
    BarG = sb.barplot(x=x, y=GoalM, ax=ax)
    BarG.set_xticklabels(BarG.get_xticklabels(), fontsize = 14)
    plt.xlabel("Month",fontsize = 18)
    plt.ylabel("Freq of Goals",fontsize = 18)
    BarG.set_title("Distribution of Monthly Goals", fontsize = 20)
    FigBar = BarG.get_figure()
  
    FigBar.savefig("BarPlot_MGoals.svg")
    
plotGoal_Month()
```
